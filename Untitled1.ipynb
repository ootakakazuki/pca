データ読み取り


```python
df = pd.read_csv("https://raw.githubusercontent.com/java-lang-programming/python_sample/master/statistics/factor_analysis/syakousei.csv")
```

```python
df
```

```python
df = df.drop('id', axis = 1)
```

相関係数

```python
df.corr()
```

```python
sns.heatmap(df.corr(), square=True, annot=True)
```

```python

# 日本語だと文字化けするため英名に変換

df = df.rename(columns={"外向性" : "Extroversion", "社交性" : "Sociability", "積極性" : "Aggressiveness", "知性" : "Intelligence", "信頼性" : "Trustworthiness", "素直さ" : "Honestly"})
df
```

```python
sns.heatmap(df.corr(), square = True, annot = True)
```

```python
plt.figure(figsize = (10, 10))
cmap = sns.color_palette("coolwarm", 200)
sns.heatmap(df.corr(), annot = True, square = True, cmap = cmap)
```

```python

# クラスターマップ
sns.clustermap(df.corr(), annot = True )
```
